```python
# K-Means Clustering on sales_data_sample.csv
# --------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Step 1: Load dataset
data = pd.read_csv("sales_data_sample.csv", encoding='latin1')

# Step 2: Select useful numeric columns
df = data[['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP']]

# Step 3: Handle missing values
df = df.dropna()

# Step 4: Standardize features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 5: Find optimal clusters using Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.show()

# Step 6: Choose optimal number of clusters (e.g., k=4 from elbow curve)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)

# Step 7: Show few results
print(df.head())

# Step 8: Visualize clusters (using first two features for simplicity)
plt.scatter(df['PRICEEACH'], df['SALES'], c=df['Cluster'], cmap='viridis')
plt.title('K-Means Clusters (PRICEEACH vs SALES)')
plt.xlabel('PRICEEACH')
plt.ylabel('SALES')
plt.show()
```
