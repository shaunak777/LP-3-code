```python
# ===========================================================
# K-Means Clustering on Sales Data (Clean Version)
# Dataset: sales_data_sample.csv (from Kaggle)
# ===========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Step 1: Load dataset
df = pd.read_csv("sales_data_sample.csv", encoding='latin1')
df.info()
df.describe()
df.isnull().sum()

# Step 2: Basic cleaning
df.drop_duplicates(inplace=True)
df = df.dropna(subset=['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP'])

# Step 3: Select numeric features
df = df[['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP']]

# Step 4: Remove outliers using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Step 5: Standardize the features
scaler = StandardScaler()
scaled = scaler.fit_transform(df)

# Step 6: Elbow method to find optimal clusters
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(7, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='blue')
plt.title('Elbow Method - Optimal Number of Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()

# Step 7: Apply K-Means with chosen k (example: 4)
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled)

# Step 8: Visualize clusters
plt.figure(figsize=(8,6))
sns.scatterplot(x='PRICEEACH', y='SALES', hue='Cluster', data=df, palette='viridis', s=60)
plt.title('K-Means Clusters (PRICEEACH vs SALES)')
plt.xlabel('PRICEEACH')
plt.ylabel('SALES')
plt.show()

# Step 9: Save results (optional)
df.to_csv('clustered_sales_data.csv', index=False)
```
